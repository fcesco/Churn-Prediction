```python
def to_datetime(date_string):
    return datetime.strptime(date_string, "%Y-%m-%d %H:%M:%S")

def column_to_datetime(df, column):
    f = udf(to_datetime, types.TimestampType())
    df = df.withColumn(column, f(df[column]))
    return df

def drop_features(stream_table, user_table):
    '''
    drop feaures hardcoded in the function
    Args:
        user_table: table with user information
        stream_table: stream with streams information
    '''

    stream_drop = ['_id', 'source_ip', 'page', 'created_at', 'title', 'text', 'message',\
                   'friends', 'place', 'secret', 'base_uri', 'media_ring', 'bucket', 'video', \
                   'audio', 'live', 'closed', 'stop_live_count', 'play_live_time', 'stop_vod_count',\
                   'play_vod_time', 'size', 'live_at', 'origin_id', 'origin_ip', 'live_confirm_time', \
                   'facebook_post_time', 'facebook_post_id', 'offline_at', 'closed_at', 'id', 'name', 'type', \
                   'public', 'likes_views', 'time', 'short_url', 'device_locale', 'file_name']
    
    user_drop = ['enabled', 'last_ip', 'latitude', 'longitude', 'fb_name', 'fb_email', \
                 'fb_access_token', 'fb_extended_access_token', 'email', 'fb_expires_at', 'fb_issued_at', \
                 'user_friends', 'latest_page', 'languages', 'device_locale', 'public_profile', 'secure_browsing',\
                 'type', 'allowed_viewers', 'facebook_id', 'created_at','bucket','fb_locale']
    
    user_table = user_table[user_table['type']=='user']
    user_table = drop_columns(user_table, user_drop)
    stream_table = drop_columns(stream_table, stream_drop)
    return stream_table, user_table

def drop_impression(impression):
    '''
    Drop not desired features from impression tables
    Arg:
        impression: impression table
    Return:
        impression table without not desired columns
    '''

    drop_list = ['time','distinct_id','app_release','carrier','city','ios_ifa','lib_version','manufacturer','model','name','os','os_version','radio','region','screen_height','screen_width','wifi','account_status','audio','broadcaster_id','broadcaster_name','facebook_post_id','page','public','screen','stream_message','type','video','facebook_id','locale','mp_country_code','mp_device_model','mp_lib','elapsed','app_version','brand']
    impression = drop_columns(impression, drop_list)
    return impression

def drop_columns(df, list_columns):
    '''
    drop multiple columns from a dataframe
    Args:
        df: Spark dataframe
        list_columns: list of columns to drop
    Return:
        df: Spark dataframe cleaned
    '''
    for column in list_columns:
        df = df.drop(column)
    return df

def import_data():
    '''
    Query data from postgres database
    Args:
        None
    Return:
        streams: Dataframe streams
        users: Dataframe users
    '''
    
    with open('access.json') as f:
        data = json.load(f)
        access_key = data['access-key']
        secret_access_key = data['secret-access-key']
        link_stream = 's3n://%s:%s@churnanalysis/stream.csv' % (access_key, secret_access_key)
        link_user = 's3n://%s:%s@churnanalysis/user.csv' % (access_key, secret_access_key)
        link_edges = 's3n://%s:%s@churnanalysis/notify.csv' % (access_key, secret_access_key)
    df_stream = sqlContext.read.load(link_stream, 
                          format='com.databricks.spark.csv', 
                          header='true', 
                          inferSchema='true')
    df_user = sqlContext.read.load(link_user, 
                      format='com.databricks.spark.csv', 
                      header='true', 
                      inferSchema='true')
    df_edges = sqlContext.read.load(link_edges, 
                  format='com.databricks.spark.csv', 
                  header='true', 
                  inferSchema='true')
    df_user.cache()
    df_stream.cache()
    df_edges.cache()
    return df_stream, df_user, df_edges

def read_impressin(impression_file):
    
    header = ['time', 'distinct_id', 'app_release', 'app_version', 'carrier',\
          'city', 'ios_ifa', 'lib_version', 'manufacturer', 'model', \
          'name', 'os', 'os_version', 'radio', 'region',  'screen_height',\
          'screen_width', 'wifi', 'account_status', 'audio', \
          'broadcaster_id', 'broadcaster_name', 'candies_sent',\
          'comments_sent', 'facebook_post_id', 'mode', 'page',  'public',\
          'screen', 'stream_duration', 'stream_id', 'stream_message', \
          'type', 'video', 'views_on_join', 'facebook_id', 'gender', \
          'locale', 'mp_country_code', 'mp_device_model', 'mp_lib',\
          'elapsed', 'app_version', 'screen_dpi', 'brand']
    fields = [StructField(field_name, StringType(), True) for field_name in\
          header]
    schema = StructType(fields)
    link_stream = 's3n://%s:%s@churnanalysis/'% (access_key, secret_access_key)
    link_impression = link_stream+a.key
    impression = sc.textFile(link_impression)
    tab_separated_impression = impression.map(lambda x: x.split('\t'))
    impression = sqlContext.createDataFrame(a, schema)
    return impression



def create_weigthed_edges(link_edges, impression_folder, user):

    edges = sqlContext.read.load(link_edges, 
                          format='com.databricks.spark.csv', 
                          header='true', 
                          inferSchema='true')
    edges.registerTempTable('old_edges')
    edges = sqlContext.sql('SELECT * FROM edges WHERE edges.exists="true"').cache()
    edges.registerTempTable('edges')
    edges = sqlContext.sql('SELECT *, 1 AS weight FROM old_edges')
    edges.registerTempTable('old_edges')
    user.registerTempTable('temp_user')
    edges = edges[(edges.follower.isin(user.facebook_id)) & \
                  (edges.followed.isin(user.facebook_id))]
    with open('access.json') as f:
        data = json.load(f)
        access_key = data['access-key']
        secret_access_key = data['secret-access-key']
    conn = boto.connect_s3(access_key, secret_access_key)
    bucket_name = 'churnanalysis'
    b = conn.get_bucket(bucket_name)
    b.list()
    filelist = []
    for file in b.list():
        if file.name[12:16] == '2016':
            filelist.append(file)
            
    user.registerTempTable('user')
    for filename in filelist:
        impression = read_impression(filename)
        impression.registerTempTable('impression')
        edges.registerTempTable('old_edges')
        edges = sqlContext.sql('SELECT old_edges.*, old_edges.weight+1 FROM old_edges \
        JOIN user ON user.facebook_id==old_edges.followed')
        
        
```
